##### 3.	A company claims that 60% of its users are satisfied with its product.In a survey of 200 users, 134 users report satisfaction. At the 5% level of significance, test whether customer satisfaction is higher than claimed.


```python
#H0 = 60%
#Ha = 134  of 200 samples so its 67 %
#level of significance = 5%
```

```python
from scipy.stats import binomtest

successes = 134      # Observed satisfied users
n_obs = 200          # Total sample size
p_null = 0.60        # Claimed proportion

result = binomtest(k=successes, n=n_obs, p=p_null, alternative='greater')

# Decision logic
alpha = 0.05
if result.pvalue < alpha:
    print("Decision: Reject the Null Hypothesis.")
else:
    print("Decision: Fail to Reject the Null Hypothesis.")
```

```python
## At 5% significance level, we reject the null hypothesis and conclude that the customer satisfaction is higher than claimed. 
#its around 67% of its users are satisfied with its product.
```
