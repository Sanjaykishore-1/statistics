```python
from scipy import stats
import numpy as np
```

```python
8.	A vaccine is tested in two regions.
In Region A, 180 out of 300 people recover.
In Region B, 165 out of 310 people recover.
Test whether the recovery rates are significantly different.

     H0 : RA  =  RB
     HA : RA /=  RB
```

```python
from statsmodels.stats.proportion import proportions_ztest

successes = np.array([180,165])
totals = np.array([300, 310])
z_stat, p_value = proportions_ztest(successes, totals, alternative='two-sided')

conv_rate_A = 180 / 300
conv_rate_B = 165 / 310

print(f"Conversion Rate A: {conv_rate_A}")
print(f"Conversion Rate B: {conv_rate_B}")
print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_value}")
```

```python
p_value = p_value/2
p_value
```

```python
alpha = 0.05
if p_value < alpha:
    print("Decision: Reject the Null Hypothesis.")
else:
    print("Decision: Fail to Reject the Null Hypothesis.")
```
