```python
9.	A company wants to compare the average productivity of employees trained using three different training programs.
At the 5% level of significance, test whether there is a difference in mean productivity across the programs.

    H0 : P1 = P2 = P3
    HA : P1/= P2/= P3
```

```python
import scipy.stats as stats

prog_1 = [18, 20, 22, 19, 21]
prog_2 = [25, 30, 28, 32, 29]
prog_3 = [20, 18, 21, 19, 22]

f_stat, p_value = stats.f_oneway(prog_1, prog_2, prog_3)

alpha = 0.05
if p_value < alpha:
    print("Conclusion: At least one group mean is significantly different.")
else:
    print("Conclusion: No significant difference between group means.")
```
