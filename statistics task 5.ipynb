##### 5.	Two teaching methods are used to train employees. A sample of 30 employees trained using Method A and 28 employees trained using Method B are tested.Test at the 5% level of significance whether there is a difference in average scores between the two methods.

```python
#H0 = Ha
#Ha /= H0 
#level of significance = 5%
```

```python
method_a = [85, 82, 91, 78, 88, 90, 89, 84, 82, 87, 92, 79, 85, 88, 83, 
            90, 81, 86, 84, 88, 90, 82, 85, 89, 87, 83, 86, 88, 81, 85]
method_b = [78, 81, 85, 76, 80, 82, 79, 83, 77, 85, 81, 80, 79, 82, 84, 
            78, 81, 83, 80, 79, 82, 85, 81, 80, 77, 79, 81, 83]

t_stat, p_value = stats.ttest_ind(method_a, method_b, equal_var=False)
p_value = p_value/2
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the Null Hypothesis (Significant difference found).")
else:
    print("Conclusion: Fail to reject the Null Hypothesis (No significant difference).")
```
