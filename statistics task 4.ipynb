#### 4.	Historically, 40% of applicants pass an entrance test. After a new preparation program, 90 out of 180 applicants pass.Test at the 5% level of significance whether the pass rate has improved.

```python
#H0 = 40%
#Ha = 90  of 180 samples so its 50 %
#level of significance = 5%
```

```python
from scipy.stats import binomtest

successes = 90      # Observed satisfied users
n_obs = 180          # Total sample size
p_null = 0.40        # Claimed proportion

result = binomtest(k=successes, n=n_obs, p=p_null, alternative='greater')

# Decision logic
alpha = 0.05
if result.pvalue < alpha:
    print("Decision: Reject the Null Hypothesis.")
else:
    print("Decision: Fail to Reject the Null Hypothesis.")
```

```python
## At 5% significance level, we reject the null hypothesis and conclude that the new preparation program has improved the pass rate of students 
#its around 50% of pass rate on new program.
```
