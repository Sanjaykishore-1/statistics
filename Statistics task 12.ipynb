```python
from scipy import stats
import numpy as np
```

```python
10. A company wants to test whether gender and product preference are independent.
A contingency table is created using survey data.
At the 5% level of significance, test for independence.
```

```python
observed = np.array([[50, 30], 
                     [40, 60]]) 

chi2_stat, p_value, dof, expected = stats.chi2_contingency(observed)


alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Gender and product preference are not independent.")
else:
    print("Fail to reject the null hypothesis: Gender and product preference are independent.")
```
