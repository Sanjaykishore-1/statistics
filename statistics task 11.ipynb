```python
from scipy import stats
import numpy as np
```

```python
11. A hospital studies whether smoking status is related to disease occurrence.
Test whether the two variables are independent.
```

```python
observed = np.array([[45, 15],  
                     [25, 65]]) 

chi2_stat, p_value, dof, expected = stats.chi2_contingency(observed)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Smoking status and disease occurrence are not independent.")
else:
    print("Fail to reject the null hypothesis: Smoking status and disease occurrence are independent.")
```
