```python
from scipy import stats
import numpy as np
```

```python
6.	The average delivery time from Warehouse A and Warehouse B is compared using random samples from both warehouses.
 At the 5% level of significance, test whether the mean delivery times differ.

we may use 2 sample t test on it 
```

```python
warehouse_a = [12.5, 13.2, 14.8, 12.9, 15.1, 14.0, 13.5]
warehouse_b = [15.2, 16.1, 14.9, 15.5, 16.8, 15.0, 15.8]

t_stat, p_value = stats.ttest_ind(warehouse_a, warehouse_b, equal_var=True)

# Decision Logic
alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject the Null Hypothesis (Significant difference found).")
else:
    print("Conclusion: Fail to Reject the Null Hypothesis (No significant difference found).")
```
