#### 2.  A company claims that the average time taken to resolve customer tickets is 30 minutes.A random sample of 40 tickets shows a mean resolution time of 33 minutes. At the 5% level of significance, test whether the average resolution time has increased.

```python
## H0 = 30 mins 
# H1 = 33 mins in 40 tickets 
# level of significance =  5% 
```

```python
average_time_taken = np.random.normal(loc=33, scale=5.0, size=40).astype(int)
```

```python
average_time_taken = np.random.normal(loc=33, scale=5.0, size=40).astype(int)
t_stats, p_value = stats.ttest_1samp(a = average_time_taken, popmean = 30, alternative='greater')
t_stats, p_value

p_value = p_value/2

if p_value > 0.05:
    print("Fail to reject the null hypothesis")
else:
    print("Reject the null hypothesis")
```

##### solution:

At 5% significance level, we reject the null hypothesis and conclude that the average time taken to resolve the tickets is significantly greater than 30 minutes.
